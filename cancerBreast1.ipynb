```python
# Creamos una columna numerica de resultado
db["diagnostico"] = db['result'].apply(lambda x : 1 if x == "M" else 0 )
```

```python
# Eliminamos la columna code del analisis
del(db['code'])
del(db['result'])
```

```python
# Ejecutamos un mapa de calor para validar las correlaciones de los campos
corr = db.corr() 
sns.heatmap(corr, annot = True)
plt.gcf().set_size_inches(50, 40)
```

```python
# Determinamos las columnas a usar en cada variable del modelo

X = db[['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z','aa','ab','ac','ad']]
y = db['diagnostico']
```

MODELO KNEIGHBORS

```python
# Ejecutamos el modelo Neighbors
# Observamos que la muestra mas certera es cuando K = 3, ya que tiene los valores mas bajos de error tanto en training como en test 

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.3, random_state=42)

for k in [1,2,3,4,5,6,10, 20]:

    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    print(f"K={k}")
    print(f"Error training: {(1 - knn.score(X_train, y_train)) * 100} %")
    print(f"Error test: {(1 - knn.score(X_test, y_test)) * 100} %")
    print("")
```

MODEL BOOSTING

```python
# definimos una funcion para el modelo de Boosting

def evaluate_model(X_train, y_train, X_test, y_test, model):
    print(f"Error training: {(1 - model.score(X_train, y_train)) * 100} %")
    print(f"Error test: {(1 - model.score(X_test, y_test)) * 100} %")
    
```

```python
# Ejecutamos el modelo Boosting
# Obtenemos con este modelo un resultado aun mas fiable que con Neighbors ya que los resultados de error son mas bajos en este modelo

from sklearn.ensemble import GradientBoostingClassifier

model = GradientBoostingClassifier(n_estimators=100, learning_rate=0.2,
    max_depth=1, random_state=0)

model = model.fit(X_train, y_train)

evaluate_model(X_train, y_train, X_test, y_test, model)
```

MODELO ARBOLES

```python
# Evaluamos un siguiente modelo a base de arboles de decision
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score

X = db[['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z','aa','ab','ac','ad']]
y = db['diagnostico']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=45)


clf = DecisionTreeClassifier(max_depth=5)
clf = clf.fit(X_train, y_train)

predictions = clf.predict(X_test)
predictions
```

```python
# Mostramos el porcentaje de acertividad del modelo
accuracy_score(y_test, predictions)
```

```python
# Evaluamos el porcentaje de precision para el diagnostico 1 = Maligno
precision_score(y_test, predictions, pos_label = 1)
```

```python
# Evaluamos el porcentaje de precision para el diagnostico 0 = Benigno
precision_score(y_test, predictions, pos_label = 0)
```
